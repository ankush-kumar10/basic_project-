#  Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
sns.set_style("whitegrid")
tips = sns.load_dataset("tips")
```

# Univariate Functions

```python
sns.histplot(data=tips, x="total_bill", bins=20, kde=True)
plt.title("Distribution of Total Bill Amounts") 
plt.show()
```

# Multiple Distributions on One Plot:


```python
sns.histplot(data=tips, x="total_bill", hue="time", multiple="dodge")
plt.title("Total Bill Distribution by Time of Day")
```

#  Basic Scatter Plot

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="time")
plt.title("Relationship between Total Bill and Tip") 
plt.show()
```

# Joint Plots with Multiple Views

```python
sns.jointplot(data=tips, x="total_bill", y="tip", kind="reg", height=8)
plt.suptitle("Joint Distribution: Total Bill vs Tip")
```

# Regression Analysis

```python
sns.regplot(data=tips, x="total_bill", y="tip", scatter_kws={'alpha':0.6}, line_kws={'color':'red'})
plt.title("Linear Regression: Total Bill vs Tip")
```

# Residual Plots

```python
sns.residplot(data=tips, x="total_bill", y="tip") 
plt.title("Residual Plot")
```

# Categorical Relationships

```python
sns.boxplot(data=tips, x="day", y="total_bill") 
plt.title("Total Bill Distribution by Day")
```

# Basic Pairplot Implementation

```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
import numpy as np  
np.random.seed(42) 
n_samples = 200 
data = { 
    'height': np.random.normal(170, 10, n_samples), 
    'weight': np.random.normal(70, 15, n_samples), 
    'age': np.random.randint(18, 80, n_samples), 
    'income': np.random.lognormal(10, 0.5, n_samples) }  
data['weight'] = data['height'] * 0.8 + np.random.normal(0, 5, n_samples) 
data['income'] = data['age'] * 1000 + np.random.normal(0, 5000, n_samples) 
df = pd.DataFrame(data)  
plt.figure(figsize=(12, 10)) 
sns.pairplot(df, diag_kind='hist', plot_kws={'alpha': 0.6}) 
plt.suptitle('Pairplot of Sample Dataset', y=1.02, fontsize=16) 
plt.show()
```

# Advanced Pairplot Technique

```python
df['category'] = np.random.choice(['A', 'B', 'C'], n_samples) 
g = sns.pairplot( df, hue='category', diag_kind='kde', plot_kws={'alpha': 0.7, 's': 30}, diag_kws={'shade': True} ) 
g.fig.suptitle('Enhanced Pairplot with Categories', y=1.02)
g.fig.set_size_inches(14, 12) 
for i, j in zip(*np.triu_indices_from(g.axes, 1)): corr = df.iloc[:, i].corr(df.iloc[:, j]) 
g.axes[i, j].annotate(f'r = {corr:.2f}', xy=(0.5, 0.5), xycoords='axes fraction', ha='center', va='center', fontsize=12, fontweight='bold') 
plt.show()
```

# Basic Correlation Heatmap

```python
correlation_matrix = df.select_dtypes(include=[np.number]).corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, square=True, fmt='.2f')
plt.title('Correlation Heatmap', fontsize=16, pad=20)
plt.tight_layout()
plt.show()
```

# Data Preparation

```python
np.random.seed(123) 
n_houses = 1000 
house_data = { 
    'sqft': np.random.normal(2000, 500, n_houses), 
    'bedrooms': np.random.randint(1, 6, n_houses), 
    'bathrooms': np.random.normal(2.5, 1, n_houses), 
    'age': np.random.randint(0, 50, n_houses), 
    'garage': np.random.randint(0, 4, n_houses) }  
house_data['price'] = ( house_data['sqft'] * 150 + house_data['bedrooms'] * 5000 + house_data['bathrooms'] * 3000 - house_data['age'] * 500 + house_data['garage'] * 2000 + np.random.normal(0, 20000, n_houses) ) 
houses_df = pd.DataFrame(house_data) 
houses_df = houses_df[houses_df['price'] > 0]
```

```python
houses_df
```

# Create Comprehensive Pairplot

```python
plt.figure(figsize=(15, 12)) 
g = sns.pairplot( houses_df, diag_kind='kde', plot_kws={'alpha': 0.6, 's': 20}, diag_kws={'shade': True} ) 
g.fig.suptitle('House Price Dataset - Pairwise Relationships', y=1.02) 
plt.show()
```

# Generate Correlation Heatmap

```python
corr_matrix = houses_df.corr()
plt.figure(figsize=(10, 8)) 
mask = np.triu(np.ones_like(corr_matrix, dtype=bool)) 
sns.heatmap( corr_matrix, mask=mask, annot=True, fmt='.3f', cmap='RdBu_r', center=0, square=True, cbar_kws={'label': 'Correlation Coefficient'} ) 
plt.title('House Price Correlation Analysis', fontsize=16, pad=20) 
plt.tight_layout() 
plt.show()
```
